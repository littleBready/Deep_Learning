# Problem 2

## Create Matrix `X`

```python
import theano
import theano.tensor as T

X = T.fmatrix('X')
```

## Create `A` and `b`

```python
from theano.tensor.shared_randomstreams import RandomStreams
from time import time

srng = RandomStreams(seed=int(time()))
A, b = srng.uniform((10, 10)), srng.uniform((10, 1))
```

## Create function and recoders

```python
from numpy import zeros

A_record, b_record = theano.shared(zeros((10, 10), dtype="Float32")), theano.shared(zeros((10, 1), dtype="Float32"))
f = theano.function([X], (X+A) * b, updates=((A_record, A_record+A), (b_record, b_record+b)))
```

## Evaluate

```python
left = f(ones((10, 10), dtype='Float32'))
right = (ones((10, 10), dtype='Float32') + A_record.get_value()) * b_record.get_value()
print left - right
```
